```python
# drop column with Id in column name
data_id = [col for col in pop_df.columns if 'Id' in col]
pop_df = pop_df.drop(data_id, axis = 1)
```

```python
# rename remaining columns to more readable names
pop_df.columns = ['county', 'pop2010all', 'pop2014all', 'pop2017all', 'pop2010_50to54','pop2014_50to54', 'pop2017_50to54', 'pop2010_55to59', 'pop2014_55to59', 'pop2017_55to59', 'pop2010_60to64', 'pop2014_60to64', 'pop2017_60to64', 'pop2010_65to69', 'pop2014_65to69', 'pop2017_65to69', 'pop2010_70to74', 'pop2014_70to74', 'pop2017_70to74', 'pop2010_75to79', 'pop2014_75to79', 'pop2017_75to79', 'pop2010_80to84', 'pop2014_80to84', 'pop2017_80to84', 'pop2010_85over', 'pop2014_85over', 'pop2017_85over', 'median_age_2010', 'median_age_2014', 'median_age_2017']
pop_df.head(2)
```

```python
# remove state name from the 'county' column
pop_df.county = pop_df.county.str.replace(', Tennessee', '')
pop_df.head(2)
```

```python
# merge pop and cancer data sets in county column
pop_and_cancer = pd.merge(pop_df, cancer_df, on = 'county', how = 'outer')
pop_and_cancer.head(2)
```

```python
# remove 'County' from merge data in county column to make data comparable
pop_and_cancer.county = pop_and_cancer.county.str.replace(' County', '')
```

```python
# create dictionary of counts of hospice facilites in each county
hospice_per_county = hospice_df.county.value_counts().to_dict()
```

```python
# see dictionary you created
hospice_per_county
```

```python
# if a county is not listed in dictionary, add the county name and assign the value 0
for county in pop_and_cancer.county:
    if county in hospice_per_county:
        continue
    else:
        hospice_per_county[county] = 0
```

```python
# now there are counties with zero hopsice facilities added to dictionary
hospice_per_county
```

```python
# create new column of the count of hospice facilities into pop_and_cancer
pop_and_cancer['hospice_count'] = [hospice_per_county[county] for county in pop_and_cancer.county]
pop_and_cancer.head()
```

```python
from sklearn import linear_model
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.model_selection import train_test_split
```

```python
X = pop_and_cancer[['hospice_count','percent_pop_over_50_2017','normalized_hospice_rate']]
Y = pop_and_cancer['5year_trend_death_rates']
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = .2, random_state = 44)
```

```python
regr = linear_model.LinearRegression()
regr.fit(X_train, Y_train)
```

```python
Y_pred = regr.predict(X_test)
```

```python
print('Coefficients: \n', regr.coef_)
print('mean squared error: %.2f'
     % mean_squared_error(Y_test, Y_pred))
print('Variance score: %.2f' % r2_score(Y_test, Y_pred))
```
